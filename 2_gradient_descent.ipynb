```python
import numpy as np
import math
import matplotlib.pyplot as plt
from IPython.display import Math

x = np.linspace(-math.pi, math.pi, 2000)
y_sin = np.sin(x)

plt.plot(x, y_sin)
```

```python
display(Math(r'Approximate: a + bx + cx^2 + dx^3'))
```

```python
a = np.random.randn()
b = np.random.randn()
c = np.random.randn()
d = np.random.randn()

def get_y_polynomial_derived():
     return a + b * x + c * x ** 2 + d * x ** 3

def plot():
    plt.plot(x, y_sin, label="Sine")
    plt.plot(x, get_y_polynomial_derived(), label="Polynomial")
    plt.legend()

plot()
```

```python
def train():
    global a,b,c,d
    learning_rate = 0.000001

    for t in range(500):
        y_polynomial_derived = get_y_polynomial_derived()

        grad_y_polynomial_derived = 2.0 * (y_polynomial_derived - y_sin)

        grad_a = grad_y_polynomial_derived.sum()
        grad_b = (grad_y_polynomial_derived * x).sum()
        grad_c = (grad_y_polynomial_derived * x ** 2).sum()
        grad_d = (grad_y_polynomial_derived * x ** 3).sum()

        descent_grad_a = learning_rate * grad_a
        descent_grad_b = learning_rate * grad_b
        descent_grad_c = learning_rate * grad_c
        descent_grad_d = learning_rate * grad_d

        a = a - descent_grad_a
        b = b - descent_grad_b
        c = c - descent_grad_c
        d = d - descent_grad_d
```

```python
train()
plot()
```
